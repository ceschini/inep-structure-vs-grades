# School structure and student's grades coorelations

Let's import both structure and grades datasets and look for correlations between them.

```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
warnings.filterwarnings('ignore')

# notebooks only
%matplotlib inline
pd.set_option('display.precision', 3)
```

```python
structure = pd.read_csv('./data/censo_inep_municipio.csv')
early_grades = pd.read_csv('./data/ideb-iniciais-municipios.csv')
later_grades = pd.read_csv('./data/ideb-finais-municipios.csv')
```

```python
structure.head()
```

```python
early_grades.head()
```

```python
def generate_correlation(df1, df2, feature, prefix):
    df_copy = df1.copy()
    df_copy[feature] = df2[feature]
    corr = df_copy.corr()
    corr.style.background_gradient(cmap='coolwarm').to_excel(
        f'./data/plots/{prefix}-{feature.lower()}-correlation.xlsx')
```

```python
generate_correlation(structure, later_grades, 'TAXA_APROVACAO', 'later')
```

```python
generate_correlation(structure, later_grades,
                     'INDICADOR_RENDIMENTO (P)', 'later')
```

```python
generate_correlation(structure, later_grades, 'NOTA_MATEMATICA', 'later')
```

```python
generate_correlation(structure, later_grades, 'NOTA_PORTUGUES', 'later')
```

```python
generate_correlation(structure, later_grades, 'NOTA_MEDIA (N)', 'later')
```

```python
generate_correlation(structure, later_grades, 'IDEB (N x P)', 'later')
```
